```python
from bokeh.io import output_notebook
from bokeh.plotting import figure,show
from bokeh.models import HoverTool
from bokeh.resources import INLINE
```

```python
output_notebook(resources = INLINE)
Tool = HoverTool(tooltips = [
    ("index", "$index"),
    ("(Watch_time_Minutes,Stream_time_minutes)", "(@Watch_time_Minutes, @Stream_time_minutes)"),
    ("Channel, Language", "@Channel, @Language")] 
)

p = figure(title = "Twitch", x_axis_label = 'Watch_time_Minutes', y_axis_label ='Stream_time_minutes', plot_width = 800, plot_height = 600, tools = [Tool])
p.circle('Watch_time_Minutes', 'Stream_time_minutes', size=2, source= th)
show(p)
```
